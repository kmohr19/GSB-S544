```python
avocados["geography"].unique()
```

```python
# make geo_type variable
avocados["geo_type"] = avocados["geography"].map({
    "Total U.S.": "US",
    "West": "region",
    "Northeast": "region",
    "South Central": "region",
    "Southeast": "region",
    "Midsouth": "region",
    "Great Lakes": "region",
    "Plains": "region",
    "California": "state",
    "South Carolina": "state"
})

# everything else is a city
avocados["geo_type"] = avocados["geo_type"].fillna("city")

avocados
```

```python
# rename columns to make more sense
avocados = avocados.rename(columns={
    "4046": "small_avocados",
    "4225": "large_avocados",
    "4770": "xl_avocados"})

avocados.columns
```

```python
# 3. Which major geographical region sold the most total organic, small Hass avocados in 2017?
organic_2017 = avocados[(avocados["type"] == "organic") & (avocados["year"] == 2017)]
organic_2017
```

```python
region_totals = (organic_2017[organic_2017["geo_type"] == "region"]
                 .groupby("geography")["small_avocados"]
                 .sum()
                 .sort_values(ascending=False))
region_totals

# the west has sold the most small avocados
```

```python
# 4. Split the date variable into month, day, and year variables. In which month is the highest average volume of avocado sales?
avocados["date"] = pd.to_datetime(avocados["date"])
avocados["year"] = avocados["date"].dt.year
avocados["month"] = avocados["date"].dt.month
avocados["day"] = avocados["date"].dt.day
avocados.columns
```

```python
avg_monthly_volume = (avocados
                      .groupby("month")["total_volume"]
                      .mean()
                      .sort_values(ascending=False))
avg_monthly_volume

# May has the highest average volume of avocado sales
```

```python
# 5. Which metro area geographical regions sold the most total avocados?
# Plot side-by-side box-plots of the total volume for only the five metro geographical regions with the highest averages for the total_volume variable

metro_data = avocados[avocados["geo_type"] == "city"]

avg_volume_by_city = (metro_data
                      .groupby("geography")["total_volume"]
                      .mean()
                      .sort_values(ascending=False)
                      .head(5))
avg_volume_by_city

```

```python
top5 = avg_volume_by_city.index.tolist()

top5_metro_data = metro_data[metro_data["geography"].isin(top5)]
top5_metro_data
```

```python
from plotnine import ggplot, geom_boxplot, aes, labs, theme_bw, theme, element_text

(ggplot(top5_metro_data, aes(x="geography", y="total_volume")) +
    geom_boxplot(fill="mediumpurple", alpha=0.7, outlier_color="gray") +
    labs(
        title="Top 5 Metro Areas by Average Total Volume",
        x="City",
        y="Total Volume of Avocados Sold")
    + theme_bw()
    + theme(
        axis_text_x=element_text(rotation=30, hjust=1),
        figure_size=(10, 6)))
```

```python
# Recreate the Plot
ca_totals = (avocado_ca
    .groupby(["geography", "type"])
    [["small_avocados", "large_avocados", "xl_avocados"]]
    .sum()
    .reset_index())
ca_totals
```

```python
ca_long = ca_totals.melt(
    id_vars=["geography", "type"],
    value_vars=["small_avocados", "large_avocados", "xl_avocados"],
    var_name="size",
    value_name="volume")
ca_long
```

```python
# proportions
proportions = pd.crosstab(
    index=[ca_long["geography"], ca_long["type"]],
    columns=ca_long["size"],
    values=ca_long["volume"],
    aggfunc="sum",
    normalize="index"
).reset_index()
proportions
```

```python
ca_long_prop = proportions.melt(
    id_vars=["geography", "type"],
    value_vars=["small_avocados", "large_avocados", "xl_avocados"],
    var_name="size",
    value_name="proportion")
ca_long_prop
```

```python
from plotnine import ggplot, aes, geom_bar, facet_wrap, labs, theme_bw, theme, scale_fill_manual, scale_y_continuous
(ggplot(ca_long_prop, aes(x="geography", y="proportion", fill="size"))
    + geom_bar(stat="identity", position="stack")
    + facet_wrap("~ type")
    + scale_fill_manual(
        values={
            "xl_avocados": "slateblue",
            "large_avocados": "mediumseagreen",
            "small_avocados": "darkorange"},
        name="Avocado Size",
        labels=["Small", "Large", "X-Large"])
    + scale_y_continuous(
        limits=(0, 1),
        breaks=[0, 0.25, 0.5, 0.75, 1],
        labels=["0%", "25%", "50%", "75%", "100%"])
    + labs(
        title="Proportion of Hass Avocado Sales by Size",
        x="California Region",
        y="Percent of Total Volume")
    + theme_bw()
    + theme(axis_text_x=element_text(rotation=30, hjust=1)))
```

```python
# Date columns into a single “date” and “home_value” column
housing_long = housing.melt(
    id_vars=["RegionName", "RegionType", "StateName"],
    var_name="date",
    value_name="home_value")
housing_long
```

```python
housing_long["date"] = pd.to_datetime(housing_long["date"], errors="coerce")
housing_long["year"] = housing_long["date"].dt.year

housing_long
```

```python
ca_metros = ["Los Angeles, CA", "San Diego, CA", "San Francisco, CA", "Sacramento, CA"]
housing_ca = housing_long[housing_long["RegionName"].isin(ca_metros)].copy()

housing_ca["RegionName"].unique()
```

```python
housing_year = (housing_ca.groupby(["RegionName", "year"], as_index=False)
              .agg(median_home_value=("home_value", "median")))

housing_year = housing_year.rename(columns={"RegionName": "geography"})

housing_year
```

```python
avocado_ca["date"] = pd.to_datetime(avocado_ca["date"])
avocado_ca["year"] = avocado_ca["date"].dt.year

avocado_ca
```

```python
avocado_ca["geography"] = avocado_ca["geography"].replace({
    "Los Angeles": "Los Angeles, CA",
    "San Diego": "San Diego, CA",
    "San Francisco": "San Francisco, CA",
    "Sacramento": "Sacramento, CA"})
avocado_ca
```

```python
avo_year = (
    avocado_ca.groupby(["geography", "type", "year"], as_index=False)
              .agg(avg_price=("average_price", "mean"),
              total_volume=("total_volume", "sum")))
avo_year
```

```python
combined_data = pd.merge(
    avo_year,
    housing_year,
    on=["geography", "year"],
    how="inner")

combined_data
```

```python
from plotnine import ggplot, aes, geom_point, geom_smooth, facet_wrap,labs, theme_bw, theme, element_text

(ggplot(combined_data, aes(x="median_home_value", y="avg_price", color="geography"))
    + geom_point(size=2, alpha=0.7)
    + geom_smooth(method="lm", se=False, linetype="dashed", size=1)
    + labs(
        title="Relationship Between Home Prices and Avocado Prices",
        x="Median Home Value",
        y="Average Avocado Price",
        color="Region")
    + facet_wrap("~ type")
    + theme_bw()
    + theme(
        axis_text_x=element_text(rotation=20, hjust=1),
        figure_size=(9,6),
        legend_position="bottom"))

```
